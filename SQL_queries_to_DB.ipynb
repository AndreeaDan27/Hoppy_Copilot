```python
import os
from dotenv import load_dotenv
import pyodbc
import sqlalchemy as sa
from sqlalchemy import create_engine

from urllib.parse import quote_plus

load_dotenv()
```

```python
#https://learn.microsoft.com/en-us/azure/azure-sql/database/connect-query-python?view=azuresql

server = os.getenv('SQL_SERVER_NAME')
database = os.getenv('SQL_DATABASE_NAME')
username = os.getenv('SQL_USER')
password = os.getenv('SQL_PASSWORD')

driver = '{ODBC Driver 17 for SQL Server}'
```

```python
conn = pyodbc.connect('DRIVER='+driver+';PORT=1433;SERVER='+server+';PORT=1443;DATABASE='+database+';UID='+username+';PWD='+ password)
cursor = conn.cursor()
```

```python
cursor.execute("SELECT TOP 3 name, collation_name FROM sys.databases")
row = cursor.fetchone()
while row:
    print (str(row[0]) + " " + str(row[1]))
    row = cursor.fetchone()
```

```python
query_get_current_database = """SELECT DB_NAME() AS CurrentDatabase"""

cursor.execute(query_get_current_database)

row = cursor.fetchone()

while row:
    print (str(row[0]))
    row = cursor.fetchone()
```

```python
select_schema = """
SELECT
    col.TABLE_NAME,
    col.COLUMN_NAME,
    col.DATA_TYPE,
    ISNULL(tc.CONSTRAINT_TYPE, 'NONE') AS CONSTRAINT_TYPE,
    col.IS_NULLABLE
FROM
    INFORMATION_SCHEMA.COLUMNS col
LEFT JOIN INFORMATION_SCHEMA.KEY_COLUMN_USAGE kcu
    ON col.COLUMN_NAME = kcu.COLUMN_NAME AND col.TABLE_NAME = kcu.TABLE_NAME
LEFT JOIN INFORMATION_SCHEMA.TABLE_CONSTRAINTS tc
    ON kcu.CONSTRAINT_NAME = tc.CONSTRAINT_NAME AND kcu.TABLE_NAME = tc.TABLE_NAME
ORDER BY
    col.TABLE_NAME,
    col.ORDINAL_POSITION;
"""

cursor.execute(select_schema)


columns = cursor.fetchall()
print("Table, and Column Details in the current database:")
for column in columns:
    print(f"""\n{column.TABLE_NAME}  
          {column.COLUMN_NAME} 
          {column.DATA_TYPE} 
          {'No Constraint' if column.CONSTRAINT_TYPE == 'NONE' else column.CONSTRAINT_TYPE}
          {'Nullable' if column.IS_NULLABLE == 'YES' else 'Not Nullable'}
        """)
```

```python
get_query = """SELECT TOP 10 * 
FROM dbo.Database_Schema"""

cursor.execute(get_query)

# Fetch and print the results
rows = cursor.fetchall()
print("First 10 rows:")
for row in rows:
    print(row)
```
